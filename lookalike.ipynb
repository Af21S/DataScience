```python
df = df.reset_index(drop=True)
knn.fit(df)  

```

```python
num_customers = 20
customer_indices = list(range(num_customers)) 
distances, indices = knn.kneighbors(df.iloc[customer_indices], n_neighbors=4)

lookalike_dict = {}

for i, customer_index in enumerate(customer_indices):
    customer_id = final_data.iloc[customer_index]['CustomerID']
    similar_customers = indices[i][1:] 
    similarity_scores = 1 - distances[i][1:] 

    lookalike_dict[customer_id] = list(zip(final_data.iloc[similar_customers]['CustomerID'], similarity_scores))


lookalike_df = pd.DataFrame.from_dict(lookalike_dict, orient='index')
lookalike_df.to_csv("Lookalike.csv", index_label="CustomerID")
```
